### The NRC Emotion Intensity Lexicon
This notebook processes the data from the [NCR Emotion Intensity Lexicon](http://saifmohammad.com/WebPages/AffectIntensity.htm), which contains a list of words and their associations with eight basic emotions (anger, anticipation, disgust, fear, joy, sadness, surprise, and trust) scored by [best--worst scalling](http://saifmohammad.com/WebPages/BestWorst.html).

```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
```

```python
# Loading the data
data = pd.read_table("/mnt/e/Desktop/EPFL/Courses/Applied_Data_Analysis/Project/NRC-Emotion-Intensity-Lexicon/NRC-Emotion-Intensity-Lexicon-v1-ForVariousLanguages-withZeroIntensityEntries.txt")
```

```python
# Filtering the data to contain only the word column and the 8 emotions with their intensities:
data_filt = data.iloc[:,0:9].copy()
```

```python
# Check the distribution of intensities for each kind of emotion in the lexicon
fig, ax = plt.subplots(2,4,figsize= (8,6), sharey = True, sharex = True)

for i in range(1,9):
 subplt = ax[i//5,i%4]
 subplt.hist(data_filt[data_filt.columns[i]][data_filt[data_filt.columns[i]] != 0])
 subplt.set_title(data_filt.columns[i])

fig.tight_layout()

fig.text(0.4,0, "Intensity")
fig.text(0,0.3, "Number of non-zero valued words", rotation = 90)
```

```python
# Calculating the number of words containing each kind of emotion
print("Number of Anger containing words: {}".format(len(data_filt.anger[data_filt.anger != 0])))
print("Number of Anticipation containing words: {}".format(len(data_filt.anticipation[data_filt.anticipation != 0])))
print("Number of Disgust containing words: {}".format(len(data_filt.disgust[data_filt.disgust != 0])))
print("Number of Fear containing words: {}".format(len(data_filt.fear[data_filt.fear != 0])))
print("Number of Joy containing words: {}".format(len(data_filt.joy[data_filt.joy != 0])))
print("Number of Sadness containing words: {}".format(len(data_filt.sadness[data_filt.sadness != 0])))
print("Number of Surprise containing words: {}".format(len(data_filt.surprise[data_filt.surprise != 0])))
print("Number of Trust containing words: {}".format(len(data_filt.trust[data_filt.trust != 0])))
```
